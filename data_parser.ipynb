```python
newfstr = ""
for msg in msgs[::-1]:
    if 'content' in msg:
        # Don't add content if it's just a react, because we are already including those
        if not ('reacted' in msg['content'] and 'to your message' in msg['content']):
            newfstr += f"{msg['sender_name']}: {format_(msg['content'])}\n"
    if 'photos' in msg:
        newfstr += f"{msg['sender_name']} sent a photo.\n"
    if 'sticker' in msg:
        newfstr += f"{msg['sender_name']} sent a sticker.\n"
    if 'audio_files' in msg:
        newfstr += f"{msg['sender_name']} sent an audio file.\n"
    if 'gifs' in msg:
        newfstr += f"{msg['sender_name']} sent a gif.\n"
    if 'share' in msg:
        newfstr += f"{msg['sender_name']} shared a link.\n"
    if 'videos' in msg:
        newfstr += f"{msg['sender_name']} shared a video.\n"
    if 'reactions' in msg:
        for r in msg['reactions']:
            newfstr += f"{r['actor']} reacted with {format_(r['reaction'])}\n"
```
